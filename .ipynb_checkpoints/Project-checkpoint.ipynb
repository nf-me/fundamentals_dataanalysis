## 1. Introduction

The Fisher Iris dataset, first introduced by Ronald Fisher in 1936, holds significant importance in the field of data science as a foundational dataset for machine learning and data analysis [3,4]. Fisher meticulously collected measurements of specific morphological features from three distinct species of Iris flowers: Iris setosa, Iris virginica, and Iris versicolor. With 50 samples per species, the dataset comprises a total of 150 samples. Fisher's detailed analysis of the morphological variations in sepal length, sepal width, petal length, and petal width provided valuable insights. These measurements were later utilised to develop a linear discriminant model for species classification [1]. The dataset's versatility spans various domains, making it a popular choice for classification and species recognition, exploratory data analysis, and comparisons/benchmarking tasks.

The comprehensive nature of the dataset, along with its historical significance and wide applicability, has made the Fisher Iris dataset a timeless resource in the data science community. It serves as an exemplary dataset for introducing fundamental concepts and techniques in machine learning and data analysis[5]. The dataset's balanced distribution, precise measurements, and distinct species categories contribute to its clarity and ease of interpretation. Researchers and educators continue to rely on the Fisher Iris dataset to teach and explore key concepts in classification, data exploration, and model evaluation[6].

With the Fisher Iris dataset as the foundation, this project aims to delve into the realm of data science by leveraging this well-known dataset to gain insights and showcase essential data analysis techniques. The project follows a systematic approach, beginning with data loading and descriptive analysis. Statistical methods are employed to identify missing values, detect outliers, analyse correlations, and assess the distribution of species within the dataset.

## 2. Classification of Variables

The dataset consists of four quantitative variables and one qualitative variable:

### Quantitative Variables (Ratio Scale):
- Sepal Length (cm)
- Sepal Width (cm)
- Petal Length (cm)
- Petal Width (cm)

### Qualitative Variable (Nominal Scale):
- Class (Iris setosa, Iris virginica, Iris versicolor)

In Python, these quantitative variables are typically represented as floating-point numbers (float64), while the qualitative variable 'Species' is best represented as a categorical type.

<a id='Summary-Statistics'></a>
## 3. Summary Statistics

For each quantitative variable, we will calculate and interpret the following summary statistics:

- Mean: Average value, providing a central tendency of the data.
- Median: Middle value, offering a better measure in the presence of outliers.
- Mode: Most frequent value, useful for categorical data.
- Standard Deviation: Measure of data spread from the mean.
- Variance: Square of standard deviation, indicating data dispersion.
- Range: Difference between maximum and minimum values

## 4. Data & Libraries

```python
# Data & libaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Importing the Penguins Dataset using CSV provided on ATU Moodle
url = "https://github.com/nf-me/fundamentals_dataanalysis/blob/master/data/iris.csv?raw=true"
df = pd.read_csv(url)

# Data types - change from object to category
df['class'] = df['class'].astype('category')

# Summary statistics
print(df.describe())
print(df['class'].mode())

# Identify the number of missing values in each column of the dataset
missing_values = df.isnull().sum()
print(missing_values)
```

## 5. Data Analysis & Visualization

### Histograms and Density Plots
- Ideal for visualizing the distribution of each quantitative variable.
- Helps in understanding the spread and skewness of data.

```python
# Set up the chart
plt.figure(figsize=(12, 8))

# Histograms for quantitative variables
quantitative_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for i, col in enumerate(quantitative_columns):
    plt.subplot(2, 2, i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Histogram of {col.replace("_", " ").capitalize()}')

plt.tight_layout()
plt.show()
```

#### Analysis and interpretation

##### Sepal Length

- The density plot peaks around the mean, indicating the most common range of sepal length. The width of the peak gives an idea of variance.


##### Sepal Width
- Similarly, the peak of the density plot aligns with the histogram bars, showing the concentration of values around the mean.


##### Petal Length and Width

- These plots also show distributions centered around their respective means, with the spread indicating variability.

### Box and Violin Plots
- Useful for identifying outliers and comparing distributions across species.
- Box plots show quartiles and medians, while violin plots provide density estimation.


```python
# Identify the outliers in the dataset (more than 3 standard deviations away from the mean).
numeric_columns = df.select_dtypes(include=['float64'])  # Select only numeric columns
outliers = ((numeric_columns - numeric_columns.mean()) / numeric_columns.std()).abs() > 3
outliers = df[outliers.any(axis=1)].dropna()
print(outliers)
```

```python
# Box Plots & Violin Plots for Numerical Variables
for col in ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']:
    plt.figure(figsize=(8, 4))

for col in ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']:
    plt.figure(figsize=(10, 6))

    # Creating a subplot for both boxplot and violin plot
    plt.subplot(1, 2, 1)
    sns.boxplot(x=col, data=df)
    plt.title(f'Box Plot of {col}')

    plt.subplot(1, 2, 2)
    sns.violinplot(x=col, data=df)
    plt.title(f'Violin Plot of {col}')

    plt.tight_layout()
    plt.show()
```

#### Analysis and interpretation

##### Sepal Length

- Box Plot: The median of sepal length is near the average, with a fairly symmetrical distribution of values in the interquartile range. There are no significant outliers, indicating a uniform spread.


- Violin Plot: Reveals a unimodal distribution with a peak around the median, suggesting most sepal lengths cluster around the central value. The plot's symmetric shape indicates a normal distribution.


##### Sepal Width

- Box Plot: Indicates a median sepal width close to the mean, with a slightly asymmetric distribution suggesting a slight skew. There might be a few outliers on the lower end, indicating some sepal widths are notably smaller than average.


- Violin Plot: Shows a more nuanced distribution, potentially bimodal, with peaks on either side of the median. This suggests two dominant groups in terms of sepal width.


##### Petal Length

- Box Plot: Exhibits a median that might be offset from the mean, with a wider interquartile range. This suggests a greater variation in petal length compared to sepal dimensions. Outliers could indicate unusually small or large petal lengths.


- Violin Plot: Likely reveals a bimodal distribution, reflecting two distinct groups of petal length. This is indicative of species variation within the dataset.


##### Petal Width

- Box Plot: Shows a median close to the mean, with a balanced spread of data within the interquartile range. The presence of outliers could indicate some exceptionally small or large petal widths.


- Violin Plot: May display a bimodal distribution, similar to petal length, which can be interpreted as a differentiation between species in terms of petal width.

### Correlation Analysis

```python
# Identify the correlation between different columns in the dataset
correlation = df.corr()
print(correlation)

# Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Correlation heatmap')
plt.tight_layout()
plt.show()
```

#### Analysis and interpretation

There is a strong positive correlation of 0.96 between petal length and petal width, indicating a linear relationship between these two variables. Similarly, sepal length and petal length show a positive correlation of 0.87, while sepal length and petal width exhibit a positive correlation of 0.81. These findings suggest that certain morphological features in the Iris flowers are closely related, potentially providing insights into their underlying biological characteristics.

### Scatter Plots & Bar Charts
- Shows relationships between two quantitative variables.
- Useful in identifying patterns, trends, and potential correlations.
- For visualising the distribution of the categorical 'class' variable.

```python
# Scatter Plot Matrix
sns.pairplot(df, hue="class", diag_kind='hist')
plt.suptitle('Pair Plot of Iris Data', y=1.02)
plt.show()

# Bar Plots for Categorical Variables
for col in ['class']:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=col, data=df)
    plt.title(f'Count Plot of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

#### Analysis and interpretation

##### Distinct Species Clustering 

- The scatter plots reveal distinct clusters, particularly when petal measurements are involved. This suggests a strong correlation between petal size and species classification.

##### Species Overlap 

- There is some overlap between species in sepal measurements, indicating these features alone might not be sufficient for accurate species classification.


##### Correlation Patterns

- Petal length and width demonstrate a clear positive correlation across all species. Sepal length and width show a more dispersed pattern, indicating a weaker correlation.

#### Bar Plot (Class Distribution)

- The count plot indicates an equal number of samples for each iris species. This uniform distribution is ideal for machine learning applications as it prevents model bias towards a particular class

### Data Distribution

```python
# Creating Test Data
def test_data(distribution, real_data, size=1000):
    if distribution == 'normal':
        return np.random.normal(np.mean(real_data), np.std(real_data), size)
    elif distribution == 'lognormal':
        return np.random.lognormal(np.mean(np.log(real_data)), np.std(np.log(real_data)), size)
    elif distribution == 'exponential':
        return np.random.exponential(np.mean(real_data), size)
    elif distribution == 'gamma':
        return np.random.gamma(np.mean(real_data), np.std(real_data), size)

# Plotting the data
def plot_comparisons(variable_name, real_data):
    plt.figure(figsize=(12, 8))
    distributions = ['normal', 'lognormal', 'exponential', 'gamma']
    
    for i, dist in enumerate(distributions, 1):
        plt.subplot(2, 2, i)
        sns.histplot(real_data, bins=30, kde=True, color='black', label='Real Data', stat='density')
        synthetic_data = test_data(dist, real_data)
        sns.kdeplot(synthetic_data, color='salmon', label=f'{dist.capitalize()} Test Data')
        plt.title(f'{variable_name} - {dist.capitalize()} Distribution')
        plt.legend()

    plt.tight_layout()
    plt.show()

# Plot for each variable in the dataset
variables = {
    'sepal_length': df['sepal_length'].dropna(),
    'sepal_width': df['sepal_width'].dropna(),
    'petal_length': df['petal_length'].dropna(),
    'petal_width': df['petal_width'].dropna()
}
```

```python
# sepal_length
plot_comparisons('sepal_length', df['sepal_length'].dropna())
```

```python
# sepal_Width
plot_comparisons('sepal_width', df['sepal_width'].dropna())
```

```python
# petal_length
plot_comparisons('petal_length', df['petal_length'].dropna())
```

```python
# petal_width
plot_comparisons('petal_width', df['petal_width'].dropna())
```

#### Analysis and interpretation

##### Sepal Measurements 

- Both sepal length and width are closely aligned with a normal distribution, indicating a symmetric variation around a mean with no significant skew.


##### Petal Measurements 

- Petal length shows a tendency towards a right-skewed distribution, making the log-normal a better fit. Petal width, however, aligns more closely with a normal distribution, indicating a balanced variation around the mean.


##### Implications 

- The normality of sepal measurements suggests that standard statistical methods can be effectively applied, while the skewness in petal length may require transformations or different modeling approaches for accurate predictions.
